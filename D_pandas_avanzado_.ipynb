```python
import pandas as pd

# Creamos un DataFrame sencillo
data = {
    "Nombre": ["Ana", "Luis", "Carlos", "Marta", "Sofía"],
    "Edad": [23, 30, 35, 28, 40],
    "Ciudad": ["Bogotá", "Medellín", "Bogotá", "Cali", "Medellín"],
    "Ventas": [100, 200, 150, 300, 250]
}
df = pd.DataFrame(data)

print(" DataFrame original:")
print(df)
```

```python
# Utilizamos el comando .replace para limpiar los datos.
print("\n Reemplazar un valor (Medellín -> MDE):")
print(df.replace({"Ciudad": {"Medellín": "MDE"}}))
```

```python
# Utilizamos el comando df para filtrar y selecionar los datos.
print("\n Filtrar personas con Edad > 30:")
print(df[df["Edad"] > 30])

print("\n Seleccionar solo Nombre y Ventas:")
print(df[["Nombre", "Ventas"]])
```

```python
# Utilizamos el comando .groupby para agrupar y scar estadisticas.
print("\n Promedio de ventas por ciudad:")
print(df.groupby("Ciudad")["Ventas"].mean())
```

```python
# Utilizamos el merge/join para unir el Df con un Df extra.
print("\n Combinar DataFrames (merge):")
df_extra = pd.DataFrame({
    "Ciudad": ["Bogotá", "Medellín", "Cali"],
    "Region": ["Centro", "Antioquia", "Pacífico"]
})
df_merged = pd.merge(df, df_extra, on="Ciudad", how="left")
print(df_merged)
```

```python
# Uitlizamos el comando .nulls para manejar valores nulos.
print("\n Crear un DataFrame con valores nulos:")
df_nulls = df.copy()
df_nulls.loc[1, "Edad"] = None
df_nulls.loc[3, "Ventas"] = None
print(df_nulls)

print("\n Detectar nulos:")
print(df_nulls.isnull())

print("\n Eliminar filas con valores nulos:")
print(df_nulls.dropna())

print("\n Rellenar nulos con un valor fijo (0):")
print(df_nulls.fillna(0))

print("\n Rellenar nulos con la media de la columna:")
print(df_nulls.fillna(df_nulls.mean(numeric_only=True)))
```

```python
# Utilizamos el .to_csv para exportar y el .read_csv para importar el Df.
print("\n Exportar a CSV:")
df.to_csv("ventas.csv", index=False)
print("Archivo 'ventas.csv' creado!")

print("\n Importar desde CSV:")
df_importado = pd.read_csv("ventas.csv")
print(df_importado)
```
