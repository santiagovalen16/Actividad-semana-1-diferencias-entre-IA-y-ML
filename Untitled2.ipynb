```python
import pandas as pd

# Creamos un DataFrame sencillo
data = {
    "Nombre": ["Ana", "Luis", "Carlos", "Marta", "Sofía"],
    "Edad": [23, 30, 35, 28, 40],
    "Ciudad": ["Bogotá", "Medellín", "Bogotá", "Cali", "Medellín"],
    "Ventas": [100, 200, 150, 300, 250]
}
df = pd.DataFrame(data)

print("📌 DataFrame original:")
print(df)
```

```python
print("\n🔹 Reemplazar un valor (Medellín -> MDE):")
print(df.replace({"Ciudad": {"Medellín": "MDE"}}))
```

```python
print("\n🔹 Filtrar personas con Edad > 30:")
print(df[df["Edad"] > 30])

print("\n🔹 Seleccionar solo Nombre y Ventas:")
print(df[["Nombre", "Ventas"]])
```

```python
print("\n🔹 Promedio de ventas por ciudad:")
print(df.groupby("Ciudad")["Ventas"].mean())
```

```python
print("\n🔹 Crear una nueva columna (Ventas * 2):")
df["Ventas_doble"] = df["Ventas"].apply(lambda x: x * 2)
print(df)
```

```python
print("\n🔹 Ejemplo con fechas:")
fechas = pd.date_range("2025-08-25", periods=5, freq="D")
df_fechas = pd.DataFrame({"Fecha": fechas, "Visitas": [10, 20, 30, 40, 50]})
print(df_fechas)

print("\n🔹 Agrupar por semana (suma de visitas):")
print(df_fechas.set_index("Fecha").resample("W").sum())
```

```python
print("\n🔹 Usar melt (columnas -> filas):")
df_melt = df.melt(id_vars=["Nombre"], value_vars=["Edad", "Ventas"])
print(df_melt)
```
